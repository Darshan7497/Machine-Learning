# Q1. Load the flight price dataset and examine its dimensions. How many rows and columns does the dataset have?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_excel('flight_price.xlsx')
df.head()
```

```python
df.shape
```

This dataset have 10683 rows and 11 columns

# Q2. What is the distribution of flight prices in the dataset? Create a histogram to visualize the distribution.

```python
# Create a histogram of flight prices
sns.histplot(df, x='Price', bins=20, edgecolor='black',kde = True)

# Add labels and title
plt.xlabel('Price')
plt.ylabel('Count')
plt.title('Distribution of Flight Prices')

# Show the plot
plt.show()
```

# Q3. What is the range of prices in the dataset? What is the minimum and maximum price?

```python
df.describe()
```

```python
# Calculate the range, minimum, and maximum price using numpy
price_range = np.ptp(df["Price"])
min_price = np.min(df["Price"])
max_price = np.max(df["Price"])

# Print the results
print(f"The range of prices is {price_range}")
print(f"The minimum price is {min_price}")
print(f"The maximum price is {max_price}")
```

# Q4. How does the price of flights vary by airline? Create a boxplot to compare the prices of different airlines.

```python
df['Airline']
```

```python
sns.boxplot(data= df, x = 'Airline',y ='Price')

# Add labels and title
plt.xlabel('Airline')
plt.ylabel('Price')
plt.title('Flight Prices by Airline')

# Show the plot
plt.show()
```

# Q14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps with the highest number of installs.

```python
# Creating a frequency table to show the number of apps in each install category
install_counts = df['Installs'].value_counts()

# Sorting the table by install counts in descending order
install_counts_sorted = install_counts.sort_values(ascending=False)

# Printing the top 10 most popular apps based on their number of installs
print(install_counts_sorted[:10])
```
