# Q1. Pearson correlation coefficient is a measure of the linear relationship between two variables. Suppose you have collected data on the amount of time students spend studying for an exam and their final exam scores. Calculate the Pearson correlation coefficient between these two variables and interpret the result.

Assuming you have collected data on n pairs of (time spent studying, exam score), you can calculate the Pearson correlation coefficient as follows:

Calculate the mean of both variables:

mean_time = (sum of all time spent studying) / n

mean_score = (sum of all exam scores) / n

Calculate the covariance between time spent studying and exam scores:

covariance = (sum of [(time_i - mean_time) * (score_i - mean_score)]) / (n - 1)

Calculate the standard deviation of both variables:

std_time = sqrt((sum of [(time_i - mean_time)^2]) / (n - 1))

std_score = sqrt((sum of [(score_i - mean_score)^2]) / (n - 1))

Compute the Pearson correlation coefficient:

r = covariance / (std_time * std_score)

The resulting value of r will range between -1 and 1. A value of -1 indicates a perfect negative linear relationship between the two variables, while a value of 1 indicates a perfect positive linear relationship. A value of 0 indicates no linear relationship between the variables.

Interpreting the result of the Pearson correlation coefficient in the context of the study, a positive value of r indicates that there is a positive linear relationship between the amount of time students spend studying for an exam and their final exam scores. In other words, as the amount of time spent studying increases, so does the final exam score. The magnitude of r indicates the strength of the relationship, with values closer to 1 indicating a stronger relationship. However, it's important to note that correlation does not imply causation and other factors could also be influencing the exam scores.

Calculate the difference between the ranks for each pair of individuals:

Calculate the Spearman's rank correlation coefficient using the formula:

r_s = 1 - (6 * sum of (d_i^2) / (n * (n^2 - 1)))
